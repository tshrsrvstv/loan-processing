```python
import h2o
import numpy as np
from sklearn.metrics import confusion_matrix, classification_report
```

```python
h2o.init(nthreads=-1)
```

```python
df = h2o.import_file('../datasets-provided/test.csv')
df.head()
```

```python
df.describe()
```

```python
df['target'] = df['target'].asfactor()
```

```python
model = h2o.load_model('../trained-models/StackedEnsemble_BestOfFamily_AutoML_20200530_103520')
```

```python
model
```

```python
y_pred = model.predict(df)
```

```python
y_pred
```

```python
print(confusion_matrix(df['target'].as_data_frame()['target'], y_pred['predict'].as_data_frame()['predict']))
```

```python
print(classification_report(df['target'].as_data_frame()['target'], y_pred['predict'].as_data_frame()['predict']))
```
